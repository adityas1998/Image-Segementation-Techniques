```python
import cv2 as cv
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
os.getcwd()
```

```python
img=cv.imread('toogreen.png',0)
```

```python
plt.imshow(img,cmap='gray',aspect='auto')
```

```python
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
ret,thresh1 = cv.threshold(img,150,255,cv.THRESH_BINARY)
plt.imshow(thresh1,'gray',aspect='auto')
```

```python
blur = cv.GaussianBlur(img,(5,5),0)
ret,th3 = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(th3,'gray',aspect='auto')
```

```python
print(np.shape(thresh1))
```

```python
thresh=[thresh1,th3]
for i in range(len(thresh)):
    for row in thresh[i]:
        cb=0
        cw=0
        for j in row:
            if j==255:
                cw+=1
            else:
                cb+=1
    if i==0:
        print("Binary\n")
    else:
        print("\nOtsu\n")
    print('Concrete=',cw/(cw+cb))
    print('Not Concrete=',cb/(cw+cb))            
```
